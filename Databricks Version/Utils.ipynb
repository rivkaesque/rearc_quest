```python
def ensure_path(Catalog,Schema,ItemName,Type,TableSchema=None):
    ##NOTE: This should only be run in minimal use cases, when typos will have a minor impact. It assumes that the Catalog already exists.
    try:
        # Create the Schema if it doesn't exist
        spark.sql(f"CREATE SCHEMA IF NOT EXISTS {Catalog}.{Schema}")
        print(f"✓ Schema {Catalog}.{Schema} ready")
        
        if Type == "table":
            # Create the Table if it doesn't exist, using TableSchema
            spark.catalog.createTable(f"{Catalog}.{Schema}.{ItemName}",schema=TableSchema,source="delta")
            print(f"✓ Volume {Catalog}.{Schema}.{ItemName} ready")
        elif Type == "volume":
            # Create the Volume if it doesn't exist
            spark.sql(f"CREATE VOLUME IF NOT EXISTS {Catalog}.{Schema}.{ItemName}")
            print(f"✓ Volume {Catalog}.{Schema}.{ItemName} ready")
        elif Type == "schema":
            print("No Volume or Table to Create")
        else:
            raise Exception(f"Error: Type {Type} not recognized")
        
    except Exception as e:
        raise Exception(f"Error: {str(e)}")
        # Todo - better error handling via logging
```
